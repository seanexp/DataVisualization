```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

# 데이터 생성하기

```python
import numpy as np
```

```python
X = np.linspace(0, 10, 10)
```

```python
X
```

```python
def f1(x):
    return x+2

def f2(x):
    return 2*x

def f3(x):
    return x**2
```

```python
plt.plot(X, f1(X), 'ro', X, f2(X), 'b-', X, f3(X), 'k--')
plt.show()
```
