```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

# 데이터 생성하기

```python
import numpy as np
```

```python
X = np.linspace(0, 10, 10)
```

```python
X
```

```python
def f1(x):
    return x+2

def f2(x):
    return 2*x

def f3(x):
    return x**2
```

```python
plt.plot(X, f1(X), 'ro', X, f2(X), 'b-', X, f3(X), 'k--')
plt.show()
```

## 축과 타이틀 쓰기

```python
plt.plot(X, f1(X), 'ro')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('tutorial')
plt.show()
```

## 다른 그래프 한꺼번에 그리기

```python
plt.subplot(211)
plt.plot(X, f2(X))
plt.title('first one')

plt.subplot(212)
plt.plot(X, f3(X))
plt.title('second one')

plt.show()
```

```python
plt.plot(X, f1(X), 'ro', label='east')
plt.plot(X, f2(X), 'b--', label='west')
plt.legend()
plt.show()
```
